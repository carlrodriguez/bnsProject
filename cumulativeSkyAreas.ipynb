# Cumulative Sky-Location Plots

The idea here is to create cumulative sky-location plots, which compare the 95% confidence intervals from our BNS runs at fixed SNR=20, to the same results from Veitch at al. 2012 (the advanced networks paper), and with Nissanke et al. 2013, where she does a similar analysis using a different MCMC code.

Necessary imports

```python
import os.path
```

The function which returns an array of sky-location areas, at the 95% credible interval, from the ranked_sky_pixels.dat produced from cbcBayesPostProc.py.  It counts up the number of pixes until it covers 95% of the credible region, then normalizes that number of pixels to the total area on the sky.  Currently it combines all the mass bins and all runs in each mass bin.

Takes as input the folder prefix, either 'HLV' or 'HLVI'

```python
def histPlots(prefix):
    masses = ['1_1/','1.4_1.4/','1_2.5/','2.5_2.5/']
    skyAreas = []

    for mass in masses:
        for i in range(41):
            skyPixels = prefix + '/' + mass + str(i) + '/post/ranked_sky_pixels.dat'
            if os.path.exists(skyPixels):
                pixel = 0
                file = open(skyPixels)
                pixelData = file.readlines()
                file.close()
                cum = float(pixelData[1].split()[3])
                while cum <= 0.95:
                    pixel += 1
                    cum = float(pixelData[pixel+1].split()[3])
                skyAreas.append(pixel*(41252.96129655765) / (len(pixelData) - 1))
                pixel = 0
    return(skyAreas)
```

Actually generate the relavant data and store it

```python
hlv = histPlots('HLV')
hlvi = histPlots('HLVI')
```

Create the cumulative histogram plot

```python
hist([hlv,hlvi],bins=50,cumulative=True,normed=True,histtype='step')
grid(True)
title("Sky-area uncertainties",fontsize=22)
ylabel("Cumulative Number",fontsize=16)
xlabel("Sky Area 95% Credible Region $(deg^2)$",fontsize=16)
```
