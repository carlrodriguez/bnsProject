# Cumulative Sky-Location Plots

The idea here is to create cumulative sky-location plots, which compare the 95% confidence intervals from our BNS runs at fixed SNR=20, to the same results from Veitch at al. 2012 (the advanced networks paper), and with Nissanke et al. 2013, where she does a similar analysis using a different MCMC code.

Necessary imports

```python
import os
```

The function which returns an array of sky-location areas, at the 95% credible interval, from the ranked_sky_pixels.dat produced from cbcBayesPostProc.py.  It counts up the number of pixes until it covers 95% of the credible region, then normalizes that number of pixels to the total area on the sky.  Currently it combines all the mass bins and all runs in each mass bin.

Takes as input the folder prefix, either 'HLV' or 'HLVI'

```python
def histPlots(prefix):
    masses = ['1_1/','1.4_1.4/','1_2.5/','2.5_2.5/']
    skyAreas = []

    for mass in masses:
        for i in range(41):
            skyPixels = prefix + '/' + mass + str(i) + '/post/ranked_sky_pixels.dat'
            if os.path.exists(skyPixels):
                pixel = 0
                file = open(skyPixels)
                pixelData = file.readlines()
                file.close()
                cum = float(pixelData[1].split()[3])
                while cum <= 0.95:
                    pixel += 1
                    cum = float(pixelData[pixel+1].split()[3])
                skyAreas.append(pixel*(41252.96129655765) / (len(pixelData) - 1))
                pixel = 0
    return(skyAreas)
```

```python
def advDetPoints():

    files = os.listdir('HLVI_AdvDet')

    skyAreas = []

    for eachFile in files:
        skyPixels = 'HLVI_AdvDet/' + eachFile
        if os.path.exists(skyPixels):
            pixel = 0
            file = open(skyPixels)
            pixelData = file.readlines()
            cum = float(pixelData[1].split()[3])
            while cum <= 0.95:
                pixel += 1
                cum = float(pixelData[pixel+1].split()[3])
            skyAreas.append(pixel*(41252.96129655765) / (len(pixelData) - 1))
            pixel = 0
    
    return(skyAreas)
```

Actually generate the relavant data and store it, including for the Advanced Detectors Paper and Nissanke 2013 results

```python
hlv = histPlots('HLV')
hlvi = histPlots('HLVI')
hlviAdvDet = advDetPoints()
NissankeHLV = np.fromfile('NissankeHLV90CI.dat',sep='\n')
NissankeHLVI = np.fromfile('Fourdet2sigmaSkyerror_India.dat',sep='\n')
```

Create a cumulative histogram plot

```python
gca().set_ylim(ymax=1.)
hist([hlviAdvDet],bins=50,cumulative=True,normed=True,histtype='step')
grid(True)
title("Sky-area uncertainties",fontsize=22)
ylabel("Cumulative Number",fontsize=16)
xlabel("Sky Area 95% Credible Region $(deg^2)$",fontsize=16)
```

Create (and save) al the plots together in a paper-ready format

```python
subplot(2,1,1)
subplots_adjust(hspace=0.4)
gca().set_ylim(ymax=1.)
gca().set_xlim(xmax=200)
hist([hlv,NissankeHLV],bins=800,cumulative=True,normed=True,histtype='stepfilled',alpha=0.8)
grid(True)
legend(['SNR$_{network}=20$', 'SNR$_{trigger}\geq 8.5$'],loc=7)
title("Sky-area uncertainties, HLV",fontsize=22)
ylabel("Cumulative Fraction",fontsize=16)
xlabel("Sky Area $95\%$ Credible Region $(deg^2)$",fontsize=16)
subplot(2,1,2)
gca().set_ylim(ymax=1.)
gca().set_xlim(xmax=200)
hist([hlvi,NissankeHLVI],bins=800,cumulative=True,normed=True,histtype='stepfilled',alpha=0.8)
grid(True)
legend(['SNR$_{network}=20$', 'SNR$_{trigger}\geq 8.5$'],loc=7)
title("Sky-area uncertainties, HLVI",fontsize=22)
ylabel("Cumulative Fraction",fontsize=16)
xlabel("Sky Area $95\%$ Credible Region $(deg^2)$",fontsize=16)
gcf().set_size_inches(6.5,9.5)
savefig('doc/comparitiveSkyAreas.pdf')
```
